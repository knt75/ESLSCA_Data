Python Lists

```python
colors = ["red", "green", "blue"]

print(colors[0])

colors[1] = "yellow"

colors.append("purple")

colors.remove("red")

print(colors)
```

Python Tuples

```python
fruits = ("apple", "banana", "cherry")

print(fruits[1])

print(len(fruits))

(a, b, c) = fruits

print(b)
```

Python Sets

```python
colors = {"red", "green", "blue"}

print(colors)

colors.add("yellow")

colors.discard("green")

print(len(colors))
```

Python Dictionaries

```python
car = {"brand": "Ford", "model": "Mustang", "year": 2024}

print(car["model"])

car["color"] = "red"

car.pop("brand")

print(car)
```

Python If...Else

```python
age = 20

if age < 13:
  print("Child")
elif age < 18:
  print("Teenager")
else:
  print("Adult")
```

Python Match

```python
day = 3

match day:
  case 3:
    print("Wednesday")
  case _:
    print("Other day")
```

Python While Loops

```python
i = 0

while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

Python For Loops

```python
fruits = ["apple", "banana", "cherry"]

for x in fruits:
  print(x)
  if x == "banana":
    break
```
